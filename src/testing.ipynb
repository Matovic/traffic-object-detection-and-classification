```python
import time

for txtFile in df['set01/video_0004/00002.txt']:
    with open((config.LABEL_DIR+txtFile), 'r') as f: # open in readonly mode
        a = f.read().split("\n") #line
        f.close()
    pom = a[0].split(" ")
    if float(pom[1])>0 and float(pom[1])<1:
        f = open((config.LABEL_DIR+txtFile), 'w')
        f.truncate(0)
        counter = 1
        for i in a:
            b = i.split(" ")
            print("Pred ",b,f,len(a),counter)#label,x,y,w,h
            if float(b[3])-0.001 > 0:
                b[3] = float(b[3])-0.001 #w
            if float(b[4])-0.001 > 0:
                b[4] = float(b[4])-0.001 #h

            print("Po: ",' '.join(str(i) for i in b))
            b =' '.join(str(i) for i in b)
            if len(a) == counter:
                f.write(b)
            else:
                f.write(b)
                f.write('\n')
            counter +=1
        f.close()

```
